## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
print(df.columns)
```

### Box Plots


Plot a box plot of `Age.`


```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'], vert=False, patch_artist=True)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import seaborn as sns
sns.set(style="whitegrid")

# Plot the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df, alpha=0.5)
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
sns.set(style="whitegrid")

# Plot the bubble plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', size='Age', data=df, sizes=(20, 200), alpha=0.7)
plt.title('Bubble Plot of WorkWeekHrs, CodeRevHrs, and Age')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
database_counts = df['DatabaseDesireNextYear'].value_counts().head(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(database_counts, labels=database_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()

```

```python
Languages_counts = df['LanguageDesireNextYear'].value_counts().head(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(Languages_counts, labels=Languages_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Languages Respondents Wish to Learn Next Year')
plt.show()
```

```python
sql_users = df['LanguageWorkedWith'].str.contains('SQL', case=False, na=False)

# Count the number of respondents working with SQL
num_sql_users = sql_users.sum()

print(f'The number of respondents working with SQL is: {num_sql_users}')
```

```python
language_worked_with = df['LanguageWorkedWith']

# Count the number of respondents working only with MySQL
num_mysql_only_users = language_worked_with.str.fullmatch('MySQL').sum()

print(f'The number of respondents working on MySQL only is: {num_mysql_only_users}')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate median WorkWeekHrs and CodeRevHrs for the age group
median_work_week_hrs = filtered_df.groupby('Age')['WorkWeekHrs'].median()
median_code_rev_hrs = filtered_df.groupby('Age')['CodeRevHrs'].median()

# Plot the stacked chart
plt.figure(figsize=(10, 6))
plt.bar(median_work_week_hrs.index, median_work_week_hrs, label='WorkWeekHrs', alpha=0.7)
plt.bar(median_code_rev_hrs.index, median_code_rev_hrs, bottom=median_work_week_hrs, label='CodeRevHrs', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.legend()
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate median ConvertedComp for each age
median_converted_comp = filtered_df.groupby('Age')['ConvertedComp'].median()

# Plot the median ConvertedComp for each age
plt.figure(figsize=(10, 6))
median_converted_comp.plot(marker='o')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()
```

```python
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]

# Calculate median ConvertedComp for each age
median_converted_comp = filtered_df.groupby('Age')['ConvertedComp'].median()

# Plot the median ConvertedComp for each age
plt.figure(figsize=(10, 6))
median_converted_comp.plot(marker='o')
plt.title('Median ConvertedComp for Ages 25 to 30')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()
```

```python
dev_type_distribution = df['MainBranch'].value_counts()

# Find the majority developer type
majority_dev_type = dev_type_distribution.idxmax()

print(f'The majority of survey responders are: {majority_dev_type}')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
main_branch_distribution = df['MainBranch'].value_counts()

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_distribution.plot(kind='barh', color='skyblue')
plt.title('Distribution of MainBranch')
plt.xlabel('Number of Responders')
plt.ylabel('MainBranch')
plt.show()

```
